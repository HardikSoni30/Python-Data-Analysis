## The `.loc()` and `.iloc()` Methods for selection

- `.loc()` is label-based, which means that we have to specify the name of the rows and columns that we want to filter out.

```python
dataframe.loc[specified rows, specified columns]
```

- `.iloc()` is index-based, which means that we have to specify rows and columns by their integer index. 

````python
selected_EmployeeNumbers = [15, 94, 337, 1120]
````

````python
data.loc[selected_EmployeeNumbers]
````

````python
data.loc[selected_EmployeeNumbers, ["Age", "Attrition"]]
````

````python
df.drop(df[mask_1 & mask_4].index, inplace=True)
df.tail()
````

````python
def replace_gender(df_fun):
    for uv in df_fun["First Name"].unique():
        subset = df_fun.loc[df["First Name"]==uv, "Gender"]
        df_fun.loc[df["First Name"]==uv, "Gender"] = subset.mode()[0]
    return df_fun
````

````python
replace_gender(df)
````

````python
df["Senior Management"] = df["Senior Management"].astype("boolean")
df["Gender"] = df["Gender"].astype("category")
df.head(10)
````

````python
mask_4 = df["Senior Management"]
mask_5 = df["Gender"] == "Male"
````

````python
df["Gender"].value_counts()
````

````python
df["Senior Management"].value_counts()
````

````python
dict_team = {
    1: df[mask_4 & mask_5]["Team"].mode()[0],
    2: df[mask_4 & ~mask_5]["Team"].mode()[0],
    3: df[~mask_4 & mask_5]["Team"].mode()[0],
    4: df[~mask_4 & ~mask_5]["Team"].mode()[0]
}
dict_team
````

````python
pd.isna(df.loc[22, "Team"]) 
````

````python
def cust_fillna_team(row):
    gender = row.iloc[1]
    sm = row.iloc[6]
    team = row.iloc[7]
    if pd.isna(team):
        if gender == "Male" and sm:
            team = dict_team.get(1)
        elif gender == "Male" and  not sm:
            team = dict_team.get(2)
        elif gender != "Male" and sm:
            team = dict_team.get(3)
        elif gender != "Male" and not sm:
            team = dict_team.get(4) 
    return team
````

````python
df[df["First Name"] == "Antonio"]
````

````python
df["Team"] = df.apply(cust_fillna_team, axis="columns")
````
